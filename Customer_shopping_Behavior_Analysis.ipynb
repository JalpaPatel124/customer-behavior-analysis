```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
labels=['Young Adult','Adult','Middle-aged','Semior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
df['frequency_of_purchases'].unique()
```

```python
#create a column purchase_frequency_days
frequency_mapping={'Fortnightly':14,
                   'Weekly':7,
                   'Annually':365,
                   'Quarterly':90, 
                   'Bi-Weekly':14,
                   'Monthly':30,
                   'Every 3 Months':90
                  }
df['frequency_mapping_days']=df['frequency_of_purchases'].map(frequency_mapping)

    
```

```python
df[['frequency_of_purchases','frequency_mapping_days']].head(10)
```
